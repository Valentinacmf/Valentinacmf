```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.size'] = 18
```

```python
TbyT0 = np.logspace(-2,1,100)

gamma = 1.2

f = 1/(gamma-1)*(TbyT0-1-gamma*np.log(TbyT0))+TbyT0-1 # ONLY WHEN p = p0

g, ax = plt.subplots(1)
ax.plot(TbyT0, f, label=r'$p/p_0 = 1$')
ax.legend(loc='lower left')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel(r'$T/T_0$ (dimensionless)')
ax.set_ylabel(r'$W_{max}/(nRT_0)$ (dimensionless)')
```
